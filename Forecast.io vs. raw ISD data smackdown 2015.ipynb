Integrating Weather Data into PE Analytics
==========================================
    
Motivation
----------
Many of the use cases for the PE Analytics project entail making inferences about variables that are plausibly effected by weather conditions. Some examples include:  

* **Equipment lifetime**  
* **Job execution time**  
* **Reducible idling time**  
  
If we can incorporate hi-resolution weather data into our models, our insights will be more accurate and we will be able to make more confident inferences.

Forecast.io
-----------
Forecast.io is application that purports to offer present and historical queries for rich weather information, plus forecasts, from any coordinates on the globe. We will evaluate this claim below. The platform incorporates data from a large variety of sources, including several from the NCDC and other sources around the globe.  

Forecast.io is accessed via a well-documented [REST API](https://developer.forecast.io/) and is free for the first 1000 calls / day. After that they start charging a fraction of a cent.  

They also make a pretty sweet iPhone app called [Dark Sky](http://darkskyapp.com/), which will ping you ten minutes before it starts raining.

NOAA / NCDC
-----------
The National Oceanic and Atmospheric Administration is really good about opening up data collected by its weather stations positioned around the country to the public. A wide variety of data products are available for consumption, from very coarse (day-level quality-controlled summaries) to very fine-grained (minute-level weather data from [ASOS](http://www.nws.noaa.gov/ost/asostech.html) stations maintained by the NWS, FAA, and DoD).  

The primary NCDC source that we have proposed to use for fine-grained US weather data is the [ISD](http://www.ncdc.noaa.gov/isd) (Integrated Surface Database). The ISD goes down to 20-minute timesteps in most places and itself incorporates data from several different types of weather stations, located thorughout the country. ISD is sort of the "Gold Standard" for NCDC weather data -- comprehensive, pretty well quality-controlled, reliable.

You can pull data from their ftp server, located at [ftp.ncdc.noaa.gov](ftp://ftp.ncdc.noaa.gov). There are usually PDFs that describe the format (sometimes arcane) of whatever you're looking at.  
  
The Questions at Hand
---------------------
* **Is Forecast.io a suitable proxy for raw ISD data?**  
  This would be nice, because Forecast.io is much easier to get at than ISD data, is pre-organized and -processed, covers more territory (the whole earth(!)) and supports real-time querying.
  
* **Do Forecast.io's interpolation strategies produce credible estimates?**
  Forecast.io's innovation is that they are able to produce a "smooth surface" describing the weather conditions over a given geo area and time. Using historical data from ISD, we can design an experiment to make sure they're not pulling our leg.  
  
* **Are Forecast.io's *forecasts* credible?**
  Using historical queries and historical ISD data, we can test Forecast.io's predictions in a few places and evaluate their performance.

So let's construct a column with the filename for each station, for easy accessing.

```python
def construct_filename(s):
    """
    Format the filename for a given station, 
    coercing types and padding with zeros where necessary.
    """
    usafid = "{0:06d}".format(int(s.usafid))  # use new python string formatting
    wban   = "{0:05d}".format(int(s.wban))
    return "{usafid}-{wban}-2015.csv".format(usafid=usafid, wban=wban)
    
# in pandas, applying a function across each value of a Series is easy.
# use the apply() method, passing it a function (like the above).
# "look ma, no for loops!"

stations['filename'] = stations.apply(construct_filename, axis=1)
stations.index = stations['filename']  # use this as the rowkey
del stations['filename']               # no need to keep the duplicate column

stations.head()
```

```python
# this ought to do for an example cluster.
# restrict to the NE corner of the map.
# here we index with a boolean array (the part inside the square brackets.)

subset = stations[stations.apply(lambda s: s.long > -103.5 and s.lat > 40.5, axis=1)]  
```

```python
# load them in
dfs = [pd.read_csv("data/csv/" + filename) for filename in subset.index]

# stack them on top of one another
df = pd.concat(dfs)

# lowercase column names
df.columns = map(str.lower, df.columns)

# make filename column for easy reference
df['filename'] = df.apply(construct_filename, axis=1)

df.head()
```

```python
# currently that df ^^ has an integer index, but it's time series data.
# let's use pandas builtin datetime indexing capabilities to make life easier.

def construct_timestamp(s):
    """
    Timestamp values are stored in different columns here; yr, m, d, hr, min. 
    We'll do a similar thing with string formatting as above, to get a standard timestamp.
    """
    return "{yr}-{m}-{d} {h}:{min}:00".format(yr=int(s["yr"]), m=int(s["m"]), d=int(s["d"]), h=int(s["hr"]), min=int(s["min"])) 

df.index = pd.DatetimeIndex(df.apply(construct_timestamp, axis=1))  # DatetimeIndex allows lots of cool query patterns.

df.head()
```

```python
# find the average temperature between 1am-2am on January 1 for station,
# and take the average among the four of them.
# hopefully, forecast.io gives a sensical value for the centroid at that time.

grouped = df['January 1, 2015 1am'].groupby("filename")
grouped.temp.aggregate(np.mean).mean()  # it's cold in northeastern Colorado!!
```

Forecast.io exposes a REST API, which also has a nice Python wrapper available on PyPi and [GitHub](https://github.com/ZeevG/python-forecast.io).

```python
# let's plot temperature for each of our four clustered stations on January 5,
# and then query Forecast.io for the "centroid point" and plot that alongside. 

day  = df["February, 2015"]
fns  = day.filename.unique()

# a dataframe with four series corresponding to temps, resampled to an hour (default strategy is taking mean)
plot_df = pd.DataFrame({str((stations.ix[fn].lat, stations.ix[fn].long)): day[day.filename == fn].temp.resample("1h") for fn in fns})
plot_df.plot()
```

```python
# define some utility functions for extracting temperature from forecastio and getting labels for stuff
def get_temp(lat, lng, datetime):
    """
    call Forecast.io API
    return temperature at the centroid at given datetime
    """
    return forecastio.load_forecast(api_key, lat, lng, time=datetime).currently().d["apparentTemperature"]

def coords(fn):
    """
    use the `stations` meta info dataframe to get coordinates for a filename.
    """
    return np.array((stations.ix[fn].lat, stations.ix[fn].long))


# datetime objects for each hour in January 5
times = [pd.datetime(2015, 1, 5, h) for h in range(24)]


# now, for each midpoint between two stations, construct a plotdf 
# containing station 1 temp data, station 2 temp data, and forecastio's temp at the midpoint.

plotdfs = []
for s1, s2 in combinations(fns, 2):
    plotdf = pd.DataFrame({str(coords(sta)): day[day.filename == sta].temp.resample("1h") for sta in (s1,s2)})

    midptlat, midptlng = tuple((coords(s1) + coords(s2)) / 2)
    plotdf["Forecast.io ({lat}, {lng})".format(lat=midptlat, lng=midptlng)] = pd.Series([get_temp(midptlat, midptlng, t) for t in times], index=plotdf.index)
    plotdfs.append(plotdf)
```

```python
for pltdf in plotdfs:
    pltdf.plot()
```
